```python
from IPython.display import HTML
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
nlp.pipe_names
```

```python
doc  = nlp("Tesla Inc acquired Twitter Inc for $45 billion.")

for ent in doc.ents:
  print(ent.text , " | " ,ent.label_ , " | ", spacy.explain(ent.label_) )
```

```python
from spacy import displacy
# Assuming 'html_content' is the HTML output generated by displacy.render
html_content = displacy.render(doc,style="ent")

# Display the HTML content directly in a Colab cell
HTML(html_content)

```

***when company is not recognized we manually set them.***

```python
doc = nlp("Tesla Inc acquired twitter for $45 billion.")
```

```python
type(doc[2:5])
```

```python
from spacy.tokens import Span
```

```python
s1 = Span(doc,0,1,label = "ORG")
s2 = Span(doc,3,4,label = "ORG")

doc.set_ents([s1,s2],default = "unmodified")
```

```python
for ent in doc.ents:
  print(ent.text, " | ",ent.label_)
```
